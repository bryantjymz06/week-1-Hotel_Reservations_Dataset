```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/raw/Hotel_Reservations.csv")
```

```python
df.head()
```

```python
# Generate descriptive statistics for the dataset #
# - lead_time has a wide range (0 to 443), meaning customers book very differently.
# - arrival_year only contains 2017, so it is not useful for prediction.
# - required_car_parking_space is mostly 0, showing low demand for parking.
df.describe()
```

```python
# checked for missing values #
# The output shows that all columns have 0 missing values.
df.isna().sum()
```

```python
##
df.corr(numeric_only=True)
```

```python
df['booking_status'] = df['booking_status'].map({'Canceled': 1, 'Not_Canceled': 0})
```

```python
## average values of each numeric feature for cancelled (1) vs not cancelled (0) bookings.
## cancelled bookings show much higher lead times and fewer repeated guests,
## indicating these features may influence cancellation likelihood.
df.groupby('booking_status').mean(numeric_only=True)
```

```python
df.corr(numeric_only=True)['booking_status'].sort_values(ascending=False)
```

```python
sns.scatterplot(x='lead_time', y='avg_price_per_room', hue='booking_status', data=df)
```
